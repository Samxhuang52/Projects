```python
import pandas as pd
import numpy as np
from pandas import DataFrame, Series
import sklearn
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('MergedWithBrightness6.8_mr.csv')
```

```python
df.dropna()
```

```python
df.info()
```

```python
lreg = LinearRegression()
```

```python
X_multi = df.drop('SecondsViewed_x',1)
X_multi = df.drop('star',1)
Y_target = df.SecondsViewed_x
```

```python
lreg.fit(X_multi,Y_target)
```

```python
print ' The estimated intercept coefficient is %.2f ' %lreg.intercept_
```

```python
print ' The number of coefficients used was %d ' % len(lreg.coef_)
```

```python
coeff_df = DataFrame(df.columns)
coeff_df.columns = ['Features']
coeff_df["Coefficient Estimate"] = pd.Series(lreg.coef_)
coeff_df
```

```python
coeff_df.sort('Coefficient Estimate')
```

```python
coeff_df.to_excel('coeff_df.xlsx')
```
